```python
import pandas as pd
```

```python
meld_df = pd.read_csv('MELD.raw/test_sent_emo.csv')
meld_df.head(4)
```

```python
meld_df = meld_df[['Utterance', 'Dialogue_ID', 'Utterance_ID', 'Emotion']]
meld_df.head(4)
```

```python
selected_emotions = ['anger', 'sadness', 'joy', 'neutral']
meld_df = meld_df[meld_df['Emotion'].isin(selected_emotions)]
```

```python
meld_df.groupby(['Emotion']).count()
```

```python
meld_df[['Utterance', 'Emotion']].groupby(['Emotion']).count().plot.bar()
```

```python
import glob
meld_df['file'] = meld_df.apply(
    lambda row : glob.glob(
        f'MELD.raw/test/output_repeated_splits_test/dia{row.Dialogue_ID}_utt{row.Utterance_ID}.mp4')[0],
    axis=1)
```

```python
import moviepy.editor as mp

for row in meld_df.iterrows():
    
    dia = row[1]['Dialogue_ID']
    utt = row[1]['Utterance_ID']
    
    text_file = f"text/dia{dia}_utt{utt}.txt"
    with open(text_file, "w", encoding="utf-8") as f:
        f.write(row[1]['Utterance'])
        
    file_path = row[1]['file']
    clip = mp.VideoFileClip(file_path)
    clip.audio.write_audiofile(f"speech/dia{dia}_utt{utt}.wav")
    
    new_clip = clip.without_audio()
    new_clip.write_videofile(f"video/dia{dia}_utt{utt}.mp4")
```

```python
meld_df['name'] = meld_df.apply(
    lambda row : f'dia{row.Dialogue_ID}_utt{row.Utterance_ID}', axis=1)
```

```python
meld_df = meld_df[['Emotion', 'name']]
```

```python
meld_df.head(5)
```

```python
labels_dict = {
    'joy' : 1, #hap
    'anger' : 0, #ang
    'sadness' : 3, #sad
    'neutral' : 2 #neu
}
meld_df['Emotion'] = meld_df['Emotion'].apply(lambda x: labels_dict[x])
meld_df.head(5)
```

```python
meld_df.to_csv('meld_test.csv', index=False)
```
